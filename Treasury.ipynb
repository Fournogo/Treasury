```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv("/Users/avery/JupyterProjects/Treasury/SITD_Sales.csv")
df["Record Date"] = pd.to_datetime(df["Record Date"])
# for column_headers in df.columns:
#     print(column_headers)

date = df[df['Security Class Description'] == 'Bond']
date = date.loc[:,"Record Date"]
y = df[df['Security Class Description'] == 'Bond']
y = y.loc[:,"Net Sales Amount"]

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(date, y)
plt.show()
```
